```python
import matplotlib.pyplot as plt
```

```python
plt.plot(oscar_movies, oscar_revenue, label="Oscar Winners")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()
plt.tight_layout()
```
