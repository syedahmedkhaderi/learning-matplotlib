```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Box Plot
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

plt.boxplot(data)
plt.grid(True)
```

```python
# Multiple Datasets

group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)

plt.boxplot([group1, group2], tick_labels=["Group1", "Group2"])
plt.show()
```
