# Components of a format string
You can combine:
- Line styles ('-', '--', '-.', ':') to control how the line is drawn.
- Markers ('o', '.', 's', '^', 'v', '*', etc.) to draw a symbol at each data point.
- Colors ('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w') for blue, green, red, cyan, magenta, yellow, black, white.


Order does not matter: "o-b", "bo-", and "-bo" all mean the same thing
